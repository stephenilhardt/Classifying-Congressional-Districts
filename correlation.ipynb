```python
import numpy as np
import pandas as pd
```

```python
sf1_numbers = pd.read_csv('sf1_numbers.csv')
sf1_percentages = pd.read_csv('sf1_percentages.csv')

acs1_numbers = pd.read_csv('acs1_numbers.csv')
acs1_percentages = pd.read_csv('acs1_percentages.csv')

social_numbers = pd.read_csv('social_numbers.csv')
social_percentages = pd.read_csv('social_percentages.csv')

economic_numbers = pd.read_csv('economic_numbers.csv')
economic_percentages = pd.read_csv('economic_percentages.csv')

housing_numbers = pd.read_csv('housing_numbers.csv')
housing_percentages = pd.read_csv('housing_percentages.csv')

predictions = pd.read_csv('predictions.csv')
```

```python
dfs = [sf1_numbers,sf1_percentages,acs1_numbers,acs1_percentages,social_numbers,social_percentages,economic_numbers,
      economic_percentages,housing_numbers,housing_percentages]

for df in dfs:
    df.drop(['Unnamed: 0'], axis=1, inplace=True)
    df['Democrat'] = predictions['Democrat']
```

```python
sf1_percentages.head()
```

```python
sf1_numbers_corr = sf1_numbers.corr()
sf1_percentages_corr = sf1_percentages.corr()

acs1_numbers_corr = acs1_numbers.corr()
acs1_percentages_corr = acs1_percentages.corr()

social_numbers_corr = social_numbers.corr()
social_percentages_corr = social_percentages.corr()

economic_numbers_corr = economic_numbers.corr()
economic_percentages_corr = economic_percentages.corr()

housing_numbers_corr = housing_numbers.corr()
housing_percentages_corr = housing_percentages.corr()
```

```python
corr_dfs = [sf1_numbers_corr,sf1_percentages_corr,acs1_numbers_corr,acs1_percentages_corr,social_numbers_corr,social_percentages_corr,economic_numbers_corr,
           economic_percentages_corr,housing_numbers_corr,housing_percentages_corr]
```

```python
len(sf1_percentages_corr)
```

```python
def make_corr_table(table): 
    df = pd.DataFrame()

    for i in range(len(table.columns)):
        app = pd.DataFrame()

        app['Column'] = table.columns
        app['Row'] = table.columns[i]
        app['Corr'] = table.lookup(app['Row'],app['Column'])
        app['Col_Dem'] = [table.loc[col,'Democrat'] for col in app['Column']]
        app['Row_Dem'] = [table.loc[row, 'Democrat'] for row in app['Row']]
        df = df.append(app)
    
    return df
```

```python
sf1_numbers_comp = make_corr_table(sf1_numbers_corr)
sf1_percentages_comp = make_corr_table(sf1_percentages_corr)

acs1_numbers_comp = make_corr_table(acs1_numbers_corr)
acs1_percentages_comp = make_corr_table(acs1_percentages_corr)

social_numbers_comp = make_corr_table(social_numbers_corr)
social_percentages_comp = make_corr_table(social_percentages_corr)

economic_numbers_comp = make_corr_table(economic_numbers_corr)
economic_percentages_comp = make_corr_table(economic_percentages_corr)

housing_numbers_comp = make_corr_table(housing_numbers_corr)
housing_percentages_comp = make_corr_table(housing_percentages_corr)
```

```python
sf1_percentages_comp.head()
```

```python
comp_dfs = [sf1_numbers_comp,sf1_percentages_comp,acs1_numbers_comp,acs1_percentages_comp,social_numbers_comp,social_percentages_comp,
            economic_numbers_comp,economic_percentages_comp,housing_numbers_comp,housing_percentages_comp]
```

```python
len(sf1_percentages_comp)
```

```python
for df in comp_dfs:
    df.reset_index(inplace=True)
    df.drop(['index'], axis=1, inplace=True)
    drop_rows = []
    for index, row in df.iterrows():
        if row[0] == row[1] or row[0] == 'Democrat' or row[1] == 'Democrat':
            drop_rows.append(index)
    df.drop(drop_rows, axis=0, inplace=True)
```

```python
for df in comp_dfs:
    df['Corr'] = np.abs(df['Corr'])
    df['Col_Dem'] = np.abs(df['Col_Dem'])
    df['Row_Dem'] = np.abs(df['Row_Dem'])
```

```python
for df in comp_dfs:
    df.dropna(inplace=True)
```

```python
len(sf1_percentages_comp)
```

```python
def remove_correlates(table):
    drop_variables = []
    
    for index, row in table.iterrows():
        if row[2] >= 0.5:
            if row[3] > row[4]:
                drop_variables.append(row[1])
            else:
                drop_variables.append(row[0])
                
    return drop_variables
```

```python
sf1_numbers_drop = remove_correlates(sf1_numbers_comp)
sf1_percentages_drop = remove_correlates(sf1_percentages_comp)

acs1_numbers_drop = remove_correlates(acs1_numbers_comp)
acs1_percentages_drop = remove_correlates(acs1_percentages_comp)

social_numbers_drop = remove_correlates(social_numbers_comp)
social_percentages_drop = remove_correlates(social_percentages_comp)

economic_numbers_drop = remove_correlates(economic_numbers_comp)
economic_percentages_drop = remove_correlates(economic_percentages_comp)

housing_numbers_drop = remove_correlates(housing_numbers_comp)
housing_percentages_drop = remove_correlates(housing_percentages_comp)
```

```python
sf1_numbers_drop = list(set(sf1_numbers_drop))
sf1_percentages_drop = list(set(sf1_percentages_drop))

acs1_numbers_drop = list(set(acs1_numbers_drop))
acs1_percentages_drop = list(set(acs1_percentages_drop))

social_numbers_drop = list(set(social_numbers_drop))
social_percentages_drop = list(set(social_percentages_drop))

economic_numbers_drop = list(set(economic_numbers_drop))
economic_percentages_drop = list(set(economic_percentages_drop))

housing_numbers_drop = list(set(housing_numbers_drop))
housing_percentages_drop = list(set(housing_percentages_drop))
```

```python
sf1_numbers_short = sf1_numbers.drop(sf1_numbers_drop, axis=1)
sf1_percentages_short = sf1_percentages.drop(sf1_percentages_drop, axis=1)

acs1_numbers_short = acs1_numbers.drop(acs1_numbers_drop, axis=1)
acs1_percentages_short = acs1_percentages.drop(acs1_percentages_drop, axis=1)

social_numbers_short = social_numbers.drop(social_numbers_drop, axis=1)
social_percentages_short = social_percentages.drop(social_percentages_drop, axis=1)

economic_numbers_short = economic_numbers.drop(economic_numbers_drop, axis=1)
economic_percentages_short = economic_percentages.drop(economic_percentages_drop, axis=1)

housing_numbers_short = housing_numbers.drop(housing_numbers_drop, axis=1)
housing_percentages_short = housing_percentages.drop(housing_percentages_drop, axis=1)
```

```python
sf1_numbers_short.drop(['Democrat'],axis=1, inplace=True)
sf1_percentages_short.drop(['Democrat'], axis=1, inplace=True)

acs1_numbers_short.drop(['Democrat'], axis=1, inplace=True)
acs1_percentages_short.drop(['Democrat'], axis=1, inplace=True)

social_numbers_short.drop(['Democrat'], axis=1, inplace=True)
social_percentages_short.drop(['Democrat'], axis=1, inplace=True)

economic_numbers_short.drop(['Democrat'], axis=1, inplace=True)
economic_percentages_short.drop(['Democrat'], axis=1, inplace=True)

housing_numbers_short.drop(['Democrat'], axis=1, inplace=True)
housing_percentages_short.drop(['Democrat'], axis=1, inplace=True)
```
